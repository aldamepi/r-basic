```python
def var(df):
  return ((df['pulso.despues']-df['pulso.antes'])/df['pulso.antes'])*100

table = run.groupby('hace.deporte').aggregate({'pulso.antes' : np.mean, 'pulso.despues': np.mean })
var(table)
```

```python
table_2 = run[run['hace.deporte'] == 'si'].groupby(
    'genero').aggregate({'pulso.antes' : np.mean, 'pulso.despues' : np.mean})
var(table_2)
```
